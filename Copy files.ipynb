# Copy Data (Monthly Data)

```python
#Dr.Dr1

# Initialize the Excel app with visible=False
app = xw.App(visible=False)

try:
    # Open the source and destination Excel files
    source_wb = xw.Book(r'Branches monthly target - 2023 Dr. Dr1.xlsx')
    destination_wb = xw.Book(r'Reports\Power BI\Dr.Dr1\Aug.xlsx')

    # Reference the sheets
    source_sheet = source_wb.sheets['Dec']
    destination_sheet = destination_wb.sheets['Sheet1']

    # Specify the source and destination cell ranges
    source_range = source_sheet['A43:G49']
    destination_range = destination_sheet['A1:G7']

    # Copy the values and paste them as values
    source_range.api.Copy()
    destination_range.api.PasteSpecial(Paste=-4163)  # -4163 corresponds to PasteValues

    # Save the destination workbook
    destination_wb.save()
    
finally:
    # Close both workbooks and quit the Excel app
    source_wb.close()
    destination_wb.close()
    app.quit()
```
